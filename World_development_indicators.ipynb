```python
mask_agr=data['IndicatorName']=='Employment in agriculture (% of total employment)'
```

```python
mask_ind=data['IndicatorName']=='Employment in industry (% of total employment)'
```

```python
mask_ser=data['IndicatorName']=='Employment in services (% of total employment)'
```

```python
mask_US=data['CountryName']=='United States'
```

```python
mask_Chi=data['CountryName']=='China'
```

```python
mask_Jap=data['CountryName']=='Japan'
```

```python
US_agr=data[mask_agr & mask_US]
print(US_agr['Year'].min(),len(US_agr) ,US_agr['Year'].max())
```

```python
US_ind=data[mask_ind & mask_US]
print(US_ind['Year'].min(),len(US_ind) ,US_ind['Year'].max())
```

```python
US_ser=data[mask_ser & mask_US]
print(US_ser['Year'].min(),len(US_ser) ,US_ser['Year'].max())
```

```python
Chi_agr=data[mask_agr & mask_Chi]
print(Chi_agr['Year'].min(),len(Chi_agr) ,Chi_agr['Year'].max())
```

```python
Chi_ind=data[mask_ind & mask_Chi]
print(Chi_ind['Year'].min(),len(Chi_ind) ,Chi_ind['Year'].max())
```

```python
Chi_ser=data[mask_ser & mask_Chi]
print(Chi_ser['Year'].min(),len(Chi_ser) ,Chi_ser['Year'].max())
```

```python
Jap_agr=data[mask_agr & mask_Jap]
print(Jap_agr['Year'].min(),len(Jap_agr) ,Jap_agr['Year'].max())
```

```python
Jap_ind=data[mask_ind & mask_Jap]
print(Jap_ind['Year'].min(),len(Jap_ind) ,Jap_ind['Year'].max())
```

```python
Jap_ser=data[mask_ser & mask_Jap]
print(Jap_ser['Year'].min(),len(Jap_ser) ,Jap_ser['Year'].max())
```

```python
plt.plot(Jap_ser['Year'].values, Jap_ser['Value'].values,label='Services %')
plt.plot(Jap_ind['Year'].values, Jap_ind['Value'].values,label='Industry %')
plt.plot(Jap_agr['Year'].values, Jap_agr['Value'].values,label='Agriculture %')
plt.xlabel('Year')
plt.ylabel('Employment share')
plt.title('Evolution of employment share in Japan')
plt.legend(loc=2)
plt.show()
```

```python
plt.plot(US_ser['Year'].values, US_ser['Value'].values,label='Services %')
plt.plot(US_agr['Year'].values, US_agr['Value'].values, label='Agriculture %')
plt.plot(US_ind['Year'].values, US_ind['Value'].values,label='Industry %')
plt.xlabel('Year')
plt.ylabel('Employment share')
plt.title('Evolution of employment share in US')
plt.legend(loc=0)
plt.show()
```
